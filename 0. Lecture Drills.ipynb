## Basic Plotting of Matplotlib

```python
# 새로운 윈도우가 아닌 현재 창에 뜨도록 함
%matplotlib inline    
import matplotlib.pyplot as plt
```

```python
plt.plot(5, 5, 'o')

plt.ylabel('y')
plt.xlabel('x')
plt.title('Plottling example')

plt.show()    # .show이후에 작성되는 코드는 plotting에 반영되지 않음
```

```python
# plt.show 이후에 작성해도 active figure를 찾아서 반영하도록 함
%matplotlib notebook    
import matplotlib.pyplot as plt
```

```python
plt.plot(5, 5, '*')
plt.show()

plt.ylabel('y')
plt.xlabel('x')
plt.title('Plotting example')
```

```python
import pandas as pd
```

```python
#1. names에 3항목인 경우 -> 3열 생성
df = pd.read_csv('immigration.txt', names = ['', 'India', 'China'])
df.drop(index=0, inplace=True)       # 첫번째 행 삭제
df.set_index('')      # 첫번째 열을 index로 옮김
```

```python
#2. names에 3항목인 경우 -> 2열 생성
import numpy as np

df = pd.read_csv('immigration.txt', names = ['India', 'China'])
df.drop(index=np.nan, inplace=True)     # 첫번째 행 삭제
df.index = df.index.astype('int64')
df
```

```python
df.plot()     # numeric data가 없다고 뜸 --> data type 확인
```

```python
df.dtypes     # 이민자 수가 object로 인식됨
```

```python
# type을 int로 변경
for x in df.columns :
    df[x] = df[x].astype('int64')    
df.dtypes  
```

```python
df.plot(kind='line')

plt.title('Immigrant comparison')
plt.xlabel('Year')
plt.ylabel('# of Immigrants')
```

```python
# Histogram
%matplotlib inline     # 기존 figure를 변경하는 오류가 발생하여 inline backend 불러옴
```
